```python
import pandas as pd
```

```python
df = pd.read_csv("hr_attrition_dirty.csv")
```

```python
print(df.head())
```

```python
# Attrition as numeric
df['Attrition_Flag'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
# Clean Age column
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
```

```python
# Remove missing ages
df = df.dropna(subset=['Age'])
```

```python
# Create Age Groups
df['Age_Group'] = pd.cut(
    df['Age'],
    bins=[18, 25, 35, 45, 60],
    labels=['18-25', '26-35', '36-45', '46-60'],
    include_lowest=True
)
```

```python
print(df[['Age', 'Age_Group']].head())
```

```python
# Clean MonthlyIncome column
df['MonthlyIncome'] = pd.to_numeric(df['MonthlyIncome'], errors='coerce')
```

```python
# Remove missing values
df = df.dropna(subset=['MonthlyIncome'])
```

```python
# Keep valid income values
df = df[df['MonthlyIncome'] > 0]
```

```python
# Create Income Bands
df['Income_Band'] = pd.cut(
    df['MonthlyIncome'],
    bins=[0, 3000, 6000, 10000, 25000],
    labels=['Low', 'Medium', 'High', 'Very High'],
    include_lowest=True
)
```

```python
print(df[['MonthlyIncome', 'Income_Band']].head())
```

```python
# Tenure group
df['Tenure_Group'] = pd.cut(df['YearsAtCompany'],
    bins=[-1, 1, 3, 7, 15, 40],
    labels=['<1 yr', '1–3 yrs', '4–7 yrs', '8–15 yrs', '15+ yrs'])
```

```python
# Satisfaction score (average of 3)
df['Avg_Satisfaction'] = (
    df['JobSatisfaction'] +
    df['EnvironmentSatisfaction'] +
    df['WorkLifeBalance']) / 3

```

```python
# High risk flag
df['High_Risk'] = (
    (df['Attrition_Flag'] == 0) &
    (df['Avg_Satisfaction'] < 2) &
    (df['OverTime'] == 'Yes') &
    (df['YearsAtCompany'] <= 3)
).astype(int)
```

```python
# Check Missing Values
print(df.isnull().sum())
```

```python
cols_to_check = [
    'Department',
    'JobRole',
    'YearsAtCompany',
    'TotalWorkingYears',
    'JobSatisfaction',
    'WorkLifeBalance',
    'PerformanceRating',
    'OverTime',
    'JoiningDate',
    'Attrition_Flag',
    'Age_Group',
    'Income_Band',
    'Tenure_Group',
    'Avg_Satisfaction'
]

df = df.dropna(subset=cols_to_check)
```

```python
print(df[cols_to_check].isnull().any(axis=1).sum())
```

```python
print(df[cols_to_check].isnull().sum())
```

```python
# Check Missing Values
print(df.isnull().sum())
```
