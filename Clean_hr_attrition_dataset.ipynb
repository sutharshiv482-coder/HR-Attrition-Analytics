```python
import pandas as pd
```

```python
df = pd.read_csv("hr_attrition_dirty.csv")
```

```python
df["department"] = df["department"].str.title()
```

```python
df["department"].unique()
```

```python
print(df.shape)
print(df.isnull().sum())
print(df.duplicated().sum())
print(df.info())
```

```python
print(df.head())
print(df.info())
print(df.describe())
print(df.shape)
```

```python
# Standardize text
df["gender"] = df["gender"].str.strip().str.lower()

# Replace variations
df["gender"] = df["gender"].replace({
    "m": "Male",
    "male": "Male",
    "f": "Female",
    "female": "Female"
})
```

```python
print(df.head())
print(df.info())
print(df.describe())
print(df.shape)
```

```python
df["attrition"] = df["attrition"].str.strip().str.title()

# Check unique values
print(df["attrition"].unique())
```

```python
df["attrition"] = df["attrition"].replace({
    "Y": "Yes",
    "N": "No",
    "YES": "Yes",
    "NO": "No"
})
```

```python
print(df.columns.tolist())
```

```python
df.columns = [
    'EmployeeID',
    'EmployeeName',
    'Age',
    'Gender',
    'Department',
    'JobRole',
    'EducationField',
    'Education',
    'MonthlyIncome',
    'HourlyRate',
    'DailyRate',
    'YearsAtCompany',
    'YearsInCurrentRole',
    'YearsWithCurrentManager',
    'TotalWorkingYears',
    'NumCompaniesWorked',
    'JobSatisfaction',
    'EnvironmentSatisfaction',
    'WorkLifeBalance',
    'PerformanceRating',
    'RelationshipSatisfaction',
    'BusinessTravel',
    'MaritalStatus',
    'DistanceFromHome',
    'OverTime',
    'PercentSalaryHike',
    'TrainingTimesLastYear',
    'StockOptionLevel',
    'JobInvolvement',
    'Attrition',
    'JoiningDate'
]
```

```python
print(df.columns.tolist())
```

```python
df["JoiningDate"] = pd.to_datetime(df["JoiningDate"], errors="coerce")
```

```python
print(df["JoiningDate"].dtype)
```

```python
df["JoiningDate"] = pd.to_datetime(
    df["JoiningDate"],
    format="%d-%m-%Y",
    errors="coerce"
)
```

```python
print(df["JoiningDate"].isna().sum())
```

```python
df["JoiningDate"] = pd.to_datetime(df["JoiningDate"], errors="coerce")

df["JoinYear"] = df["JoiningDate"].dt.year
df["JoinMonth"] = df["JoiningDate"].dt.month_name()
df["JoinQuarter"] = df["JoiningDate"].dt.quarter
```
